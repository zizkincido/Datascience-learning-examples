```python
#for predict models
df_accidents_categorical = df_accidents.drop(['Road_Type','Light_Conditions','Weather_Conditions','Road_Surface_Conditions','Time'],axis=1)
df_accidents_categorical
```

```python
X = df_accidents_categorical.drop(["Accident_Severity"],axis=1).values
y = df_accidents_categorical["Accident_Severity"].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(X_train, y_train)
randomforest_y_pred = regr.predict(X_test)
randomforest_mse = mean_squared_error(y_test, randomforest_y_pred)
```

# Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train)
gnb_y_pred = gnb.predict(X_test)
gnb_mse = mean_squared_error(y_test, gnb_y_pred)
```

# K-Neighbor

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=2)
neigh.fit(X_train, y_train)
neigh_y_pred=neigh.predict(X_test)
neigh_mse = mean_squared_error(y_test, neigh_y_pred)
```

# Decision Tree Regression

```python
from sklearn import tree

decreg = tree.DecisionTreeRegressor()
decreg = decreg.fit(X_train, y_train)
decreg_y_pred=decreg.predict(X_test)
decreg_mse=mean_squared_error(y_test,decreg_y_pred)
```
