```python
df1.drop(['Region','Standard Error','Family','Dystopia Residual'],axis=1,inplace=True)
df2.drop(['Region','Lower Confidence Interval','Upper Confidence Interval','Family','Dystopia Residual'],axis=1,inplace=True)
df3.drop(['Whisker.high','Whisker.low','Family','Dystopia.Residual'],axis=1,inplace=True)
df4.drop(['Social support'],axis=1,inplace=True)
df5.drop(['Social support'],axis=1,inplace=True)
```

```python
print(df1.columns)
print(df2.columns)
print(df3.columns)
print(df4.columns)
print(df5.columns)
```

```python
df1 = df1.rename(columns={'Economy (GDP per Capita)': 'GDP per Capita', 'Health (Life Expectancy)': 'Healthy life expectancy','Trust (Government Corruption)':'Government Corruption'})
df1 = df1.assign(Year='2015')
 
df2 = df2.rename(columns={'Economy (GDP per Capita)': 'GDP per Capita', 'Health (Life Expectancy)': 'Healthy life expectancy','Trust (Government Corruption)':'Government Corruption'})
df2 = df2.assign(Year='2016')

df3 = df3.rename(columns={'Happiness.Rank': 'Happiness Rank','Happiness.Score':'Happiness Score','Economy..GDP.per.Capita.': 'GDP per Capita', 'Health..Life.Expectancy.': 'Healthy life expectancy','Trust..Government.Corruption.':'Government Corruption','Dystopia.Residual':'Dystopia Residual'})
df3 = df3.assign(Year='2017')

df4 = df4.rename(columns={'Overall rank': 'Happiness Rank','Country or region':'Country','Score':'Happiness Score','Freedom to make life choices':'Freedom','Perceptions of corruption':'Government Corruption','GDP per capita':'GDP per Capita'})
df4 = df4.assign(Year='2018')

df5 = df5.rename(columns={'Overall rank': 'Happiness Rank','Country or region':'Country','Score':'Happiness Score','Freedom to make life choices':'Freedom','Perceptions of corruption':'Government Corruption','GDP per capita':'GDP per Capita'})
df5 = df5.assign(Year='2019')
print(df1.columns)
print(df2.columns)
print(df3.columns)
print(df4.columns)
print(df5.columns)
```

```python
df = pd.concat([df1,df2,df3,df4,df5])
df
```

```python
#Most Happy Countries By Year
df[df['Happiness Rank'] <= 10]
```

```python
df.describe()
```

```python
groups = df.groupby('Country')


for name, group in groups:
  std = group['Happiness Score'].std()
  mask = group['Happiness Score'].abs().gt(std)
  filtered_group = group[mask]


filtered_group
```
